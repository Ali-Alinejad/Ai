```python
import numpy as np
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4]).reshape(-1, 1)
y=np.array([1, 3, 5, 7]).reshape(-1, 1)
x
```

```python
plt.plot(x, y , 'r+')
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```

```python
m = model.coef_[0]
c = model.intercept_
print('slope:', m, 'intercept:', c)
```

```python
plt.plot(x, y , 'r+')
plt.plot(x,model.predict(x), 'b-')
```

```python
r = model.predict([[5]])
r
```
